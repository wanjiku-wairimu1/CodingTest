```python
import geopandas as gpd
from shapely.geometry import Polygon

# Define the bounding box of Africa (longitude and latitude limits)
africa_bbox = (-20, -35, 55, 40)  # (min_lon, min_lat, max_lon, max_lat)

# Define the grid cell size (100km x 100km)
cell_size_km = 100

# Create an empty GeoDataFrame to store the grid cells
grid_gdf = gpd.GeoDataFrame(columns=['geometry'])

# Iterate over longitude and latitude ranges to create grid cells
for lon in range(int(africa_bbox[0]), int(africa_bbox[2]), cell_size_km):
    for lat in range(int(africa_bbox[1]), int(africa_bbox[3]), cell_size_km):
        # Define the corners of the grid cell
        cell_poly = Polygon([
            (lon, lat),
            (lon + cell_size_km, lat),
            (lon + cell_size_km, lat + cell_size_km),
            (lon, lat + cell_size_km),
        ])
        # Append the cell to the GeoDataFrame
        grid_gdf = grid_gdf.append({'geometry': cell_poly}, ignore_index=True)

# Set the GeoDataFrame's coordinate reference system (CRS)
grid_gdf.crs = 'EPSG:4326'  # WGS 84

# Save the grid as a GeoJSON file
grid_gdf.to_file("africa_grid.geojson", driver='GeoJSON')

print("Grid generation completed and saved as africa_grid.geojson.")
```

```python
import geopandas as gpd
from shapely.geometry import Polygon

# Define the bounding box of Africa (longitude and latitude limits)
africa_bbox = (-20, -40, 60, 40)  # (min_lon, min_lat, max_lon, max_lat)

# Define the cell size (100km x 100km)
cell_size_deg = 1.0  # 1 degree is approximately 111 km

# Create an empty GeoDataFrame to store the grid cells
grid_gdf = gpd.GeoDataFrame(columns=['geometry'])

# Iterate over longitude and latitude ranges to create grid cells
for lon in range(int(africa_bbox[0]), int(africa_bbox[2]), int(cell_size_deg * 2)):
    for lat in range(int(africa_bbox[1]), int(africa_bbox[3]), int(cell_size_deg * 2)):
        # Define the corners of the grid cell
        cell_poly = Polygon([
            (lon, lat),
            (lon + (cell_size_deg * 2), lat),
            (lon + (cell_size_deg * 2), lat + (cell_size_deg * 2)),
            (lon, lat + (cell_size_deg * 2)),
        ])
        # Append the cell to the GeoDataFrame
        grid_gdf = grid_gdf.append({'geometry': cell_poly}, ignore_index=True)

# Set the GeoDataFrame's coordinate reference system (CRS)
grid_gdf.crs = 'EPSG:4326'  # WGS 84

# Save the grid as a GeoJSON file
grid_gdf.to_file("africa_grid.geojson", driver='GeoJSON')

print("Grid generation completed and saved as africa_grid.geojson.")

```
